```python
from pylab import *
from pandas import *
import matplotlib.pyplot as plt
import numpy as np
ts = Series(randn(1000) # ガウス分布でランダムな値をとる（numpy）
            , index=date_range( # Dateのデータを生成（pandas）
                               '2000-01-01' # スタートの日付
                               , periods=1000)  # 個数
           )
ts = ts.cumsum() # 累積和
ts.plot()
plt.show()
```

```python
# -*- coding: utf-8 -*-
from pylab import *
from pandas import *
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager # 日本語を使うために必要
fontprop = matplotlib.font_manager.FontProperties(fname="C:\Windows\Fonts/ipaexg.ttf") # フォントファイルの場所を指定
df = DataFrame(randn(1000, 4), # 1000行4列のデータ
               index=ts.index, # Series で使ってたデータのインデックス
               columns=list('ABCD') # 列名の指定
              )
df.hist(color='b', alpha=0.5, bins=50)
df = df.cumsum() # 累積和を計算
# print(df)
plt.figure() # ファイル保存の単位になるらしい（これやると新しくなるのかな）
df.plot()
#plt.legend(loc='best')
ax = df.plot(secondary_y=['A', 'B'])
ax.set_ylabel('CD の売上', fontdict = {"fontproperties": fontprop}) # 日本語フォントを指定
ax.right_ax.set_ylabel('AB スケール', fontdict = {"fontproperties": fontprop}) # 日本語フォントを指定
# plt.show()
# plt.savefig("image2.png")
```

```python
df.plot(subplots=True, figsize=(6, 6)) # 別々にプロットする
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
df['A'].plot(ax=axes[0,0])
axes[0,0].set_title('A')
df['B'].plot(ax=axes[0,1])
axes[0,1].set_title('B')
df['C'].plot(ax=axes[1,0])
axes[1,0].set_title('C')
df['D'].plot(ax=axes[1,1])
axes[1,1].set_title('D')
```

```python
df.diff().hist(color='b', alpha=0.5, bins=50) # 前の値との差を取る（累積を元に戻す）
```
