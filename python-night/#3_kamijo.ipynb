```python
import math
import numpy as np
from matplotlib import pyplot

pi = math.pi   #mathモジュールのπを利用

x = np.linspace(0, 2*pi, 100)  #0から2πまでの範囲を100分割したnumpy配列
y = np.sin(x)

#pyplot.plot(x, y)
#凡例のためにlabelキーワードで凡例名をつける
pyplot.plot(x, y, label='sin')

#グラフタイトル
pyplot.title('Sin Graph')

#グラフの軸
pyplot.xlabel('X-Axis')
pyplot.ylabel('Y-Axis')

#グラフの凡例
pyplot.legend()

pyplot.show()
```

```python
pi = math.pi

x = np.linspace(0, 10, 1000)
sin_y = np.sin(x)
cos_y = np.cos(x)  #新たにcosを計算

pyplot.plot(x, sin_y, label='sin')
pyplot.plot(x, cos_y, label='cos')  #cosの値をプロット

#グラフタイトル
pyplot.title('Sin And Cos Graph')

#グラフの軸
pyplot.xlabel('X-Axis')
pyplot.ylabel('Y-Axis')

#グラフの凡例
pyplot.legend()

pyplot.show()
```

```python
x = np.linspace(0, 10, 1000)
y = np.exp(x)

#凡例のためにlabelキーワードで凡例名をつける
pyplot.plot(x, y, label='exp')

#グラフタイトル
pyplot.title('Exponential Graph')

#グラフの軸
pyplot.xlabel('X-Axis')
pyplot.ylabel('Y-Axis')

#グラフの凡例
pyplot.legend()

#pyplot.show()
```

```python
# データ生成
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# プロット領域(Figure, Axes)の初期化
fig = pyplot.figure(figsize=(15, 8))
ax1 = fig.add_subplot(241)
ax2 = fig.add_subplot(242)
ax3 = fig.add_subplot(243)
ax5 = fig.add_subplot(245)
ax6= fig.add_subplot(246)
ax8= fig.add_subplot(248)

# 棒グラフの作成
ax1.bar([1,2,3],[3,4,5])
ax2.barh([0.5,1,2.5],[0,1,2])
ax3.scatter(y1, y2)

# 水平線、垂直線を入れる
ax3.axhline(0)
ax3.axvline(0)
ax5.axhline(0.5)
ax5.axvline(0.5)

pyplot.show()
```

```python
x = np.linspace(-10, 10, 100)
y = np.sqrt(2*x*x) + x + 1

pyplot.plot(x, y)
pyplot.show()
```

```python
# -*- coding: utf-8 -*-

def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y = sigmoid(x)

pyplot.plot(x, y)
pyplot.axhline(y=1,color='r', linestyle='--')
pyplot.axhline(y=0.5,color='black', linestyle='--')
pyplot.axhline(y=0,color='r', linestyle='--')
pyplot.axvline(x=0,color='black', linestyle='-')

#グラフタイトル
pyplot.title('シグモイド関数')
pyplot.show()
```
